```python
df_left = df.copy()
df_left['date'] = df_left['date'] + pd.Timedelta(days = 1)
```

```python
df_diff = df.merge(df_left[['date', 'fips', 'cases', 'deaths']], on = ['date', 'fips'], how = 'left')
```

```python
df_diff['cases_y'] = df_diff['cases_y'].fillna(0)
df_diff['deaths_y'] = df_diff['deaths_y'].fillna(0)
df_diff['cases'] = df_diff['cases_x'] - df_diff['cases_y']
df_diff['deaths'] = df_diff['deaths_x'] - df_diff['deaths_y']
df_diff = df_diff.drop(columns = ['cases_x', 'cases_y', 'deaths_x', 'deaths_y'])
```

```python
df_diff.head()
```

```python
#add day_of_week field
df_diff['day_of_week']=df_diff['date'].dt.dayofweek   
```

```python
#convert day_of_week to Int
df_diff['day_of_week'] =df_diff['day_of_week'].astype(int)
#rename day_of_week field 
days = {0:'0_Mon',1:'1_Tues',2:'2_Weds',3:'3_Thurs',4:'4_Fri',5:'5_Sat',6:'6_Sun'}
df_diff['day_of_week'] = df_diff['day_of_week'].apply(lambda x: days[x])
df_diff['month']=pd.DatetimeIndex(df_diff['date']).month
```

```python
us_holiday= []
for ptr in holidays.US(years = [2020, 2021,2022]).items():
    us_holiday.append(ptr)

#convert to DataFrame    
us_holiday = pd.DataFrame(us_holiday, columns = ['holiday_dt', 'holiday_nm'])
us_holiday['holiday_dt'] = pd.to_datetime(us_holiday['holiday_dt'], format = '%Y-%m-%d')
```

```python
us_holiday.head()
```

```python
df_diff_holi = pd.merge(left=df_diff, right=us_holiday, how='left', left_on='date', right_on='holiday_dt')
df_diff_holi=df_diff_holi.drop(columns=['holiday_dt'])
```

```python
df_diff_holi.head()
```

```python
pd.pivot_table(df_diff_holi, values='deaths', 
                                index=['month'], 
                                columns=['day_of_week'], 
                                aggfunc=np.mean,
                                ).plot.bar(figsize=(10,7),
                                           title='Mean deaths count by month and day of week')
```

```python
pd.pivot_table(df_diff_holi, values='cases', 
                                index=['month'], 
                                columns=['day_of_week'], 
                                aggfunc=np.mean,
                                ).plot.bar(figsize=(10,7),
                                           title='Mean cases count by month and day of week')
```

```python
pd.pivot_table(df_diff_holi, values='deaths', 
                                index=['holiday_nm'], 
                                aggfunc=np.mean,
                                ).plot.bar(figsize=(10,7),
                                           title='Mean deaths count by US holidays')
```
