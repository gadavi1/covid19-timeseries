```python
df_left = df.copy()
df_left['date'] = df_left['date'] + pd.Timedelta(days = 1)
```

```python
df_diff = df.merge(df_left[['date', 'fips', 'cases', 'deaths']], on = ['date', 'fips'], how = 'left')
```

```python
df_diff['cases_y'] = df_diff['cases_y'].fillna(0)
df_diff['deaths_y'] = df_diff['deaths_y'].fillna(0)
df_diff['cases'] = df_diff['cases_x'] - df_diff['cases_y']
df_diff['deaths'] = df_diff['deaths_x'] - df_diff['deaths_y']
df_diff = df_diff.drop(columns = ['cases_x', 'cases_y', 'deaths_x', 'deaths_y'])
```

```python
df_target = df_diff.copy()[0:0]
```

```python
df_target_denom = pd.DataFrame()
for x in df_diff['date'].unique():
    df_diff_temp = df_diff.loc[df_diff['date'] <= x]

    sorted_val = df_diff_temp.loc[df_diff_temp['cases'] > 0].sort_values(by='date')
    recent_result_cases = sorted_val.drop_duplicates('fips', keep='last')

#    sorted_val = df_diff_temp.loc[df_diff_temp['deaths'] > 0].sort_values(by='date')
#    recent_result_deaths = sorted_val.drop_duplicates('fips', keep='last')

    df_diff_temp = df_diff_temp.loc[df_diff_temp['date'] == x]

    df_cases_target = df_diff_temp.merge(recent_result_cases[['date', 'fips', 'cases', 'deaths']], on = ['fips'], how = 'left')
    df_cases_target['days_since_gt0'] = df_cases_target['date_x'] - df_cases_target['date_y']
    df_cases_target.rename(columns={'date_x':'date', 'cases_y':'recent_cases_gt0'}, inplace=True)
    df_cases_target = df_cases_target.drop(columns = ['date_y', 'cases_x', 'deaths_x', 'deaths_y'])
    df_target_denom = df_target_denom.append(df_cases_target)
```

```python
df_left = df_diff.copy()
df_left['date'] = df_left['date'] - pd.Timedelta(days = 1)
df_lookforward_01 = df_diff.merge(df_left[['date', 'fips', 'cases', 'deaths']], on = ['date', 'fips'], how = 'inner')
df_lookforward_01.rename(columns={'cases_y':'cases', 'deaths_y':'deaths'}, inplace=True)
df_lookforward_01 = df_lookforward_01.drop(columns = ['cases_x', 'deaths_x'])
```

```python
df_target_denom = df_target_denom.dropna()
df_cases_completion = df_target_denom.merge(df_lookforward_01[['date', 'fips', 'cases', 'deaths']], on = ['date', 'fips'], how = 'inner')
df_cases_completion['_target'] = df_cases_completion['cases'] / df_cases_completion['recent_cases_gt0']
df_cases_completion.loc[(df_cases_completion['_target'] < 0 ),'_target'] = 0
#Double log transform
df_cases_completion['_target'] = np.log1p(df_cases_completion['_target'])
df_cases_completion['_target'] = np.log1p(df_cases_completion['_target'])
```
